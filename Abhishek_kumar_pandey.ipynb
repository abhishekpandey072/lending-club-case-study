
# Those who are going to default we will analyse them and only can be done for the case for fully paid or charged off loans.

Removing the records with current loan status
For the "current loans" we cannot use for our anaylysis as they are already approved.

#  Missing values Check

```python
(loan_data_set.isna().sum()/len(loan_data_set.index))*100
```


# Handling Missing values
- columns with missing values:-
    "emp_length", "revol_util"
- Lets check what kind of data each column has before mixing above columns.

```python
loan_data_set.info()
```

```python
print("Mode : " + loan_data_set.emp_length.mode()[0])
loan_data_set.emp_length.value_counts()
```

#  The Mode value has more higher frequency than that of the next most frequent value as per the above value counts .
- This means that we can safely assign the value of mode to the null values in the column.
- We have imputed missing values with Mode aas they are in very low percentage hence, we it won't affect much on our analysis

```python
loan_data_set.emp_length.fillna(loan_data_set.emp_length.mode()[0], inplace = True)
loan_data_set.emp_length.isna().sum()
```

```python
loan_data_set.dropna(axis = 0, subset = ['revol_util'] , inplace = True)
loan_data_set.revol_util.isna().sum()
```

# Making data standard
- "revol_util" column although described as an object column, it has continous values.
- So we need to standardize the data in this column
- "int_rate" is one such column.
- "emp_length" have values like --> { (< 1 year) is assumed as 0 and 10+ years is assumed as 10 }
- Although the datatype of "term" is arguable to be an integer, there are only two values in the whole column and it might as  well be declared a categorical variable.

```python
loan_data_set.revol_util = pd.to_numeric(loan_data_set.revol_util.apply(lambda x : x.split('%')[0]))
```

```python
loan_data_set.int_rate = pd.to_numeric(loan_data_set.int_rate.apply(lambda x : x.split('%')[0]))
```

```python
loan_data_set.emp_length = pd.to_numeric(loan_data_set.emp_length.apply(lambda x: 0 if "<" in x else (x.split('+')[0] if "+" in x else x.split()[0])))
```

```python
loan_data_set.head()
```

# We can clearly obseve in above boxplot which is indicating the presence of outliers.

- There is huge increase in the value for small quantile variation
- The values after 95 percentile seems to be disconected from the general distribution and also .
